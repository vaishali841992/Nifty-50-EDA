# What is Nifty 50
The NIFTY 50 index is National Stock Exchange of India's benchmark broad based stock market index for the Indian equity market. NIFTY 50 stands for National Index Fifty, and represents the weighted average of 50 Indian company stocks in 17 sectors. It is one of the two main stock indices used in India, the other being the BSE Sensex.

# Import Libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt #data Visualization
import seaborn as sns #data Visualization
import plotly.io as pio
```

# Data Set

```python
df=pd.read_csv('NIFTY_50.csv')
df
```

# Lets underastand the data, We have these columsÂ¶
* symbol = stocks names of nifty 50
* Open=Open price for current day
* High=Price made high for current day
* Low=Price made low for current day
* Chng=Change in rupees
* % Chng=Percentage of change from previous day close
* Volume (lacs)=Traded volume
* Turnover (crs.)=Turnover
* 52w H=Stock 52 week high price
* 52w L=Stock 52 week low price
* 365 d % chng=Stock price percentage change from one year
* 30 d % chng=Stock price percentage change from last 30days

# DataTypes

```python
df.dtypes
```

# Type Casting

```python
df['Open'] = df['Open'].astype(str).str.replace(',', '').astype(float)
df['High'] = df['High'].astype(str).str.replace(',', '').astype(float)
df['Low'] = df['Low'].astype(str).str.replace(',', '').astype(float)
df['LTP'] = df['LTP'].astype(str).str.replace(',', '').astype(float)
df['Turnover (crs.)'] = df['Turnover (crs.)'].astype(str).str.replace(',', '').astype(float)
df['Volume (lacs)'] = df['Volume (lacs)'].astype(str).str.replace(',', '').astype(float)
df['52w H'] = df['52w H'].astype(str).str.replace(',', '').astype(float)
df['52w L'] = df['52w L'].astype(str).str.replace(',', '').astype(float)
```

```python
for col in num_cols:
    print("Column : ", col)
    print(f"Value counts of {col}: ",df[col].value_counts().sum())
    print(f"Number of Unique Values in {col}: ",df[col].nunique())
    print()
```

```python
for cols in str_cols:
    print("Column : ", cols)
    print(f"Value counts of {cols}: ",df[cols].value_counts().sum())
    print(f"Number of Unique Values in {cols}: ",df[cols].nunique())
    print()
```

# Bar plot

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation=90)


sns.barplot(data = df, x = 'Symbol', y = '30 d % chng')
plt.show
```

* Plotting BarPlot, we can find that majority of the companies have shown negative returns if stayed invested for a month in it.
* Nestle is showing down coz there was corporate action in share i.e there is drastic change in stock
* Hence investment period should be more than a month.

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation=90)


sns.barplot(data = df, x = 'Symbol', y = '365 d % chng')
plt.show
```

 * Plotting BarPlot, we can find that majority of the companies have shown atleast 10% returns if stayed invested for a year in it

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation=90)


sns.barplot(data = df, x = 'Symbol', y = 'Chng')
plt.show
```

```python
df[(df['Chng']>0)]

plt.figure(figsize=(15,7))
plt.xticks(rotation=90)


sns.barplot(data = df[(df['Chng']>20)], x = 'Symbol', y = 'Chng')
plt.show
```

 * Above are the comapnies which have given more than 20 % returns

```python
sort_yearly=df.sort_values('365 d % chng')
sort_yearly
```

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation=90)


sns.barplot(data = sort_yearly, x = 'Symbol', y = '365 d % chng')
plt.xlabel("Symbol")  
plt.ylabel('% Change in a Year(Sorted)')  
plt.title("Yearly Growth(Sorted)")  
plt.show
```

 * BPCL gave the highest returns in a year, more than 30%
* Hindalco gave the lowest returns in a year -10 %

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation=90)


sns.lineplot(data = sort_yearly, x = 'Symbol', y = '365 d % chng')
plt.xlabel("Symbol")  
plt.ylabel('% Change in a Year(Sorted)')  
plt.title("Yearly Growth(Sorted)")  
plt.show
```
